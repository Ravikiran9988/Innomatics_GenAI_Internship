Problem 1: Smart Parking Lot Management System

```python
def smart_parking(capacity, logs):
    current = 0   # Current vehicles inside parking
    peak = 0      # Track maximum vehicles at any time

    # Loop through each vehicle log
    for log in logs:
        if log == "IN":
            current += 1   # Vehicle entered
        elif log == "OUT" and current > 0:
            current -= 1   # Vehicle exited

        # Update peak usage if current exceeds previous peak
        if current > peak:
            peak = current

    # Display results
    print("Currently Parked Vehicles:", current)
    print("Peak Parking Usage:", peak)

    # Check if capacity exceeded
    if current > capacity:
        print("Parking Status: Exceeded Capacity ⚠")
    else:
        print("Parking Status: Available")


# Sample Run
smart_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

Problem 2: Online Food Delivery Time Estimator

```python
def estimate_delivery(distance, traffic, weather):
    # Base time calculation (5 minutes per km)
    base_time = distance * 5

    # Add delay based on traffic condition
    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 10
    else:
        base_time += 5

    # Add delay based on weather condition
    if weather == "Rainy":
        base_time += 10
    elif weather == "Storm":
        base_time += 20

    # Final ETA
    print("Estimated Delivery Time:", base_time, "minutes")


# Sample Run
estimate_delivery(8, "High", "Rainy")
```

Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
def seat_occupancy(total_seats, booked_seats):
    # Count total booked seats using list length
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", round(occupancy), "%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Open Additional Show")
    else:
        print("Show Status: Seats Available")


# Sample Run
seat_occupancy(200, [1] * 150)
```

Problem 4: Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    # Calculate average CPU usage
    average = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", round(average), "%")

    # Classify based on rules
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Run
classify_server_load([45, 60, 70, 85, 90])
```

Problem 5: Smart Classroom Resource Usage Monitor

```python
def resource_monitor(usage):
    overused = []  # List to store overused resources

    # Loop through dictionary items
    for resource, hours in usage.items():
        # If usage exceeds 8 hours, consider overused
        if hours > 8:
            overused.append(resource)

    # Display result
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Run
resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

Problem 6: Online Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):

    # Check if registrations exceed capacity
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    # Display final registration details
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Run
event_registration(100, 105)
```
