```python
# 1.User Login Check

correct_username = "admin"
correct_password = "1234"

user = input("Enter username: ")
pwd = input("Enter password: ")

if user == correct_username and pwd == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# 2.Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_students = 0
fail_students = 0

for mark in marks:
    if mark >= 50:
        pass_students += 1
    else:
        fail_students += 1

print("Pass students:", pass_students)
print("Fail students:", fail_students)
```

```python
# 3.Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

clean_names = []

for n in names:
    n = n.strip()      # remove spaces from start/end
    n = n.lower()      # convert to lowercase
    clean_names.append(n)

print("\nCleaned names:", clean_names)
```

```python
# 4.Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    msg_len = len(msg)
    print("\nMessage:", msg)
    print("Length:", msg_len)

    if msg_len > 10:
        print("This message is longer than 10 characters")
```

```python
# 5.Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_total = 0

for item in logs:
    if item == "ERROR":
        error_total += 1

print(error_total)
```
