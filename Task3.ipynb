Problem 1: Employee Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# highest score
highest_score = 0

for score in employees.values():
    if score > highest_score:
        highest_score = score

# who got highest score
top_employees = []

for name in employees:
    if employees[name] == highest_score:
        top_employees.append(name)

print("Top Performers Eligible for Bonus:", end=" ")
print(", ".join(top_employees), "(Score:", highest_score, ")")
```

Problem 2: Search Query Analysis

```python
query = "Buy mobile phone buy phone online"

# Convert everything to lowercase
query = query.lower()

# Split sentence into words
words = query.split()

word_frequency = {}

# Count how many times each word appears
for word in words:
    if word in word_frequency:
        word_frequency[word] = word_frequency[word] + 1
    else:
        word_frequency[word] = 1

# Show only words that appear more than once
result = {}

for word in word_frequency:
    if word_frequency[word] > 1:
        result[word] = word_frequency[word]

print(result)
```

Problem 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_data = []

# Go through each reading
for i in range(len(sensor_readings)):
    reading = sensor_readings[i]

    # Check if the number is even
    if reading % 2 == 0:
        valid_data.append((i, reading))

print("Valid Sensor Readings (Hour, Value):")
print(valid_data)
```

Problem 4:Email Domain Usage

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Extract domain and count
for email in emails:
    parts = email.split("@")
    domain = parts[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total = len(emails)

# Calculate percentage
for domain in domain_count:
    percentage = (domain_count[domain] / total) * 100
    print(domain + ":", int(percentage), "%")
```

Problem 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Find average sales
total_sales = 0

for amount in sales:
    total_sales += amount

average_sales = total_sales / len(sales)

# 30% above average
limit = average_sales + (0.30 * average_sales)

# Check which days crossed the limit
for i in range(len(sales)):
    if sales[i] > limit:
        print("Day", i + 1, ":", sales[i])
```

Problem 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count_ids = {}

# Count each user ID
for user in user_ids:
    if user in count_ids:
        count_ids[user] = count_ids[user] + 1
    else:
        count_ids[user] = 1

# Show only duplicates
for user in count_ids:
    if count_ids[user] > 1:
        print(user, "→", count_ids[user], "times")
```
